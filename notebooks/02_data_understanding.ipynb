### 1. Helper function for profiling
We will write a reusable function `profile_dataframe` to generate statistical summaries, check for duplicates, count unique values, examine memory footprints, and map null rates for each column in a structured format.

````python
def profile_dataframe(df, name="Dataset"):
    print("=" * 60)
    print(f"PROFILING: {name}")
    print("=" * 60)
    print(f"Shape (Rows, Columns): {df.shape}")
    print(f"Duplicate Rows: {df.duplicated().sum()}")
    print(f"Memory Usage: {df.memory_usage(deep=True).sum() / (1024**2):.2f} MB")
    print("\n--- Column-wise Metadata ---")
    
    cols_summary = []
    for col in df.columns:
        null_count = df[col].isnull().sum()
        null_pct = (null_count / len(df)) * 100
        unique_count = df[col].nunique()
        dtype = df[col].dtype
        sample = df[col].dropna().iloc[0] if len(df[col].dropna()) > 0 else np.nan
        
        cols_summary.append({
            "Column": col,
            "Dtype": dtype,
            "Null Count": null_count,
            "Null Pct (%)": round(null_pct, 2),
            "Unique Values": unique_count,
            "Sample Value": sample
        })
        
    summary_df = pd.DataFrame(cols_summary)
    display(summary_df)
    
    print("\n--- Statistical Overview (Numerical & Categorical) ---")
    display(df.describe(include='all'))
    print("\n")
````

### 3. Detailed Data Dictionary

Below is the documentation of columns across all tables. This maps keys, datatypes, and descriptions.

#### 3.1. Customers Table (`olist_customers_dataset.csv`)
- **`customer_id`** (object, PK): Key linking to `orders`. It is unique per order, not per customer.
- **`customer_unique_id`** (object): Unique identifier for the customer (used for customer retention & RFM analysis).
- **`customer_zip_code_prefix`** (int64, FK): First 5 digits of customer zip code (links to `geolocation`).
- **`customer_city`** (object): Customer's city name.
- **`customer_state`** (object): Customer's state code (e.g., SP, RJ).

#### 3.2. Geolocation Table (`olist_geolocation_dataset.csv`)
- **`geolocation_zip_code_prefix`** (int64, PK/FK): First 5 digits of zip code.
- **`geolocation_lat`** (float64): Latitude coordinates.
- **`geolocation_lng`** (float64): Longitude coordinates.
- **`geolocation_city`** (object): City name.
- **`geolocation_state`** (object): State code.

#### 3.3. Order Items Table (`olist_order_items_dataset.csv`)
- **`order_id`** (object, FK): Unique identifier of the order.
- **`order_item_id`** (int64, PK): Sequential number identifying the number of items included in the same order.
- **`product_id`** (object, FK): Product identifier.
- **`seller_id`** (object, FK): Seller identifier.
- **`shipping_limit_date`** (object): Date and time by which the seller must hand over the product to the logistics partner.
- **`price`** (float64): Selling price per item.
- **`freight_value`** (float64): Shipping cost per item.

#### 3.4. Order Payments Table (`olist_order_payments_dataset.csv`)
- **`order_id`** (object, FK): Unique identifier of the order.
- **`payment_sequential`** (int64, PK): Index starting from 1 for orders paid with multiple methods.
- **`payment_type`** (object): Payment method (credit_card, debit_card, voucher, boleto).
- **`payment_installments`** (int64): Number of installment payments selected.
- **`payment_value`** (float64): Total transaction value of this payment segment.

#### 3.5. Order Reviews Table (`olist_order_reviews_dataset.csv`)
- **`review_id`** (object, PK): Unique review identifier.
- **`order_id`** (object, FK): Unique identifier of the order reviewed.
- **`review_score`** (int64): Rating from 1 (worst) to 5 (best).
- **`review_comment_title`** (object): Customer title of review comments (in Portuguese).
- **`review_comment_message`** (object): Customer review comment text (in Portuguese).
- **`review_creation_date`** (object): Review collection date.
- **`review_answer_timestamp`** (object): Review respondent answer timestamp.

#### 3.6. Orders Table (`olist_orders_dataset.csv`)
- **`order_id`** (object, PK): Unique identifier of the order.
- **`customer_id`** (object, FK): Key linking to the `customers` table (identifying the specific order transaction).
- **`order_status`** (object): State of the order (delivered, shipped, canceled, invoiced, processing, approved, created, unavailable).
- **`order_purchase_timestamp`** (object): Purchase date and time.
- **`order_approved_at`** (object): Date and time payment was approved.
- **`order_delivered_carrier_date`** (object): Date/time order was picked up by carrier.
- **`order_delivered_customer_date`** (object): Date/time order was delivered to customer.
- **`order_estimated_delivery_date`** (object): Date/time estimated delivery by Olist.

#### 3.7. Products Table (`olist_products_dataset.csv`)
- **`product_id`** (object, PK): Unique identifier of the product.
- **`product_category_name`** (object, FK): Category name in Portuguese.
- **`product_name_lenght`** (float64): Character length of product name.
- **`product_description_lenght`** (float64): Character length of description.
- **`product_photos_qty`** (float64): Number of photos in product listing.
- **`product_weight_g`** (float64): Weight in grams.
- **`product_length_cm`** (float64): Length in cm.
- **`product_height_cm`** (float64): Height in cm.
- **`product_width_cm`** (float64): Width in cm.

#### 3.8. Sellers Table (`olist_sellers_dataset.csv`)
- **`seller_id`** (object, PK): Unique identifier of the seller.
- **`seller_zip_code_prefix`** (int64, FK): First 5 digits of seller zip code.
- **`seller_city`** (object): Seller city name.
- **`seller_state`** (object): Seller state code.

#### 3.9. Product Category Translation Table (`product_category_name_translation.csv`)
- **`product_category_name`** (object, PK): Category name in Portuguese.
- **`product_category_name_english`** (object): Equivalent category name in English.

### 5. Critical Observations & Operational Assumptions
Based on initial data definitions, we note key characteristics:
1. **Customer ID Differentiator**: `customer_id` is an order-level identifier. A customer placing multiple orders will have a separate `customer_id` for each order, but their overall customer identity is captured by `customer_unique_id` in the `customers` table. Any repeat customer or LTV calculations MUST group by `customer_unique_id`.
2. **Multiple Payments**: `order_payments` can have multiple rows per order (sequential indices 1, 2, 3, etc.) when split payments are used (e.g. paying part with a voucher and part with a credit card).
3. **Multiple Order Items**: An order can contain multiple rows in `order_items`. Items in the same order may come from different sellers, which means shipment dispatch dates and logistics carrier updates are tracked at the item/seller level.
4. **Geolocations**: The `geolocation` table contains high cardinality and duplicates because multiple coordinates can exist for a single zip code prefix. We will need to aggregate/de-duplicate this table before mapping state/city routes to avoid cartesian joins.
