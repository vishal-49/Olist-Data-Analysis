# Olist Brazilian E-Commerce Business Analysis
## Phase 2: Business Understanding & Problem Statement

![Olist Banner](https://raw.githubusercontent.com/olist/olist-dataset/master/olist.png)

---

### 1. Business Background
**Olist** is a prominent Brazilian SaaS (Software as a Service) technology company operating in the e-commerce space. It acts as an **integrator/marketplace hub**, allowing small and medium-sized enterprises (SMEs) across Brazil to list and sell their products directly on major retail channels (e.g., Americanas, Submarino, Shoptime, Casas Bahia, Mercado Livre) through a single Olist account.

Once a merchant uploads their products to Olist:
1. The products are listed on various partner marketplaces under the Olist brand name.
2. Customers buy the products on these marketplaces.
3. Olist notifies the seller to pack and ship the item using Olist's logistics partners (e.g., Correios, private carriers).
4. The customer reviews the purchase, and Olist collects reviews, payments, and coordinates customer service.
5. Olist payouts the seller, minus commission fees.

Because Olist represents the face of the transaction to the end-consumer, **operational deficiencies (such as delayed shipments, poor product quality, or sub-par customer service) directly damage Olist's reputation and search ranking on major partner marketplaces.**

---

### 2. Business Problem Statement
Despite rapid scaling and high transaction volumes, Olist faces operational frictions that threaten its growth, merchant retention, and customer satisfaction. Specifically:
- **Logistics Frictions**: Long shipping delays, high cancellation rates, and inaccurate estimated delivery dates lead to customer dissatisfaction and increased operational overhead.
- **Seller Quality Control**: Uneven seller performance, where a subset of sellers repeatedly delivers late or ships low-quality items, leading to high chargebacks and negative reviews.
- **Customer Retention**: E-commerce operates on tight margins where customer acquisition costs (CAC) are high. The business currently lacks clear visibility into customer lifetime value (CLV) and repeat-purchase behaviors.
- **Payment Optimization**: Understanding how payment methods (Boleto, Credit Card, Vouchers) and installments affect order size (AOV) and default rates is critical to optimize checkout conversions and merchant cash flow.

---

### 3. Business Objectives
The primary strategic objectives of this project are:
1. **Maximize Revenue Growth**: Identify high-performing categories, states, and sellers to target marketing and inventory replenishment.
2. **Enhance Customer Satisfaction (NPS)**: Pinpoint root causes of low review scores and operationalize strategies to raise average ratings.
3. **Optimize Logistics Operations**: Analyze shipping performance, find transit bottlenecks, and propose realistic estimated delivery dates.
4. **Increase Customer Lifetime Value**: Establish RFM customer segmentation to drive target marketing, loyalty programs, and remarketing campaigns.
5. **Improve Seller Compliance**: Design a quantitative Seller Scorecard to identify, reward, or penalize merchants based on compliance and review scores.

---

### 4. Key Performance Indicators (KPIs)
To track progress towards these objectives, we define the following KPIs:

| KPI Category | Metric Name | Definition | Business Target |
| :--- | :--- | :--- | :--- |
| **Financial** | **Gross Merchandise Value (GMV)** | Total value of products sold on the platform | Increase by 15% YoY |
| **Financial** | **Average Order Value (AOV)** | Total Revenue / Total Orders | Increase by 5% through cross-selling |
| **Customer** | **Average Review Score** | Mean score of reviews (1 to 5 stars) | Maintain > 4.2 out of 5.0 |
| **Logistics** | **Late Delivery Rate (LDR)** | % of orders delivered after the estimated date | Decrease below 5.0% |
| **Logistics** | **Average Delivery Time** | Mean days from purchase to customer delivery | Reduce to < 10 days nationwide |
| **Retention** | **Repeat Purchase Rate (RPR)** | % of customers with > 1 completed orders | Increase from current baseline by 3% |

---

### 5. Stakeholders
- **Executive Board (C-Suite)**: Monitising strategy, company growth, and platform-wide NPS.
- **Logistics & Supply Chain Team**: Directing carrier contracts, warehousing solutions, and shipping route optimization.
- **Merchant Relations / Seller Support**: Managing onboarding policies, quality standards, and seller commission metrics.
- **Marketing & CRM Team**: Running promotional campaigns, loyalty rewards, and customer segment targeting.
- **Finance Team**: Credit card processor negotiations, installments financing, and Boleto discount schemes.

---

### 6. Key Business Questions to Answer
Through this data analytics pipeline, we will answer the following business questions:

#### A. Revenue & Growth
1. What is the monthly trend of GMV and transaction volume? Is sales growth showing seasonality?
2. Which product categories generate the maximum revenue and order volume? Which should receive inventory focus?
3. Which Brazilian states and cities represent the core revenue drivers? Where is the highest growth potential?

#### B. Customer & Loyalty
4. What proportion of customers are repeat buyers, and what is their contribution to total GMV?
5. What is the distribution of Customer Lifetime Value (CLV) and how can we segment customers using RFM (Recency, Frequency, Monetary)?

#### C. Logistics & Operations
6. What is the average delivery delay, and which states experience the worst shipping delays?
7. Does delivery delay directly impact review scores? What is the threshold delay that triggers 1-star reviews?
8. What is the percentage of late deliveries, and are estimated delivery dates too optimistic or too pessimistic?

#### D. Seller & Product Management
9. How is seller performance distributed? What share of sellers is responsible for the majority of complaints?
10. Can we create a Seller Scorecard based on revenue, fulfillment speed, and ratings to optimize seller quality?

#### E. Payments & Checkout
11. Which payment methods are most popular? Do customers who pay with installments spend more per order (higher AOV)?
12. What is the impact of payment type on order cancellation rate?

---

### 7. Expected Business Impact
By answering these questions and deploying our analytics solution, Olist will be equipped to:
- **Reduce churn** of high-value customers by targeting them with personalized loyalty offers.
- **Save logistics cost** by identifying bottlenecks and renegotiating SLAs with underperforming carrier companies in slow regions.
- **Boost NPS** by proactively notifying customers about delays before they happen, and handling poor-performing sellers.
- **Drive sales conversion** by customizing payment and checkout options based on purchasing power and installment trends.
